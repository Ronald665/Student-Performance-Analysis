```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Base_data = pd.read_csv('StudentsPerformance.csv')
```

```python
Base_data_copy = Base_data.copy()
```

```python
Base_data_copy['Average'] = Base_data_copy.Average.round(1)
```

```python
Base_data_copy.head()
```

```python
Base_data_copy['math score'].describe()
```

```python
sns.kdeplot(data = Base_data_copy, x = 'math score', fill = True)
plt.title('General Performance in Mathematics')
```

```python
sns.countplot(data = Base_data_copy, x = 'Math Group')
plt.title('General Performance in Mathematics')
plt.xticks(fontsize = 7)
plt.show()
```

```python
Base_data_copy.gender.value_counts()
```

```python
sns.countplot(data = Base_data_copy, x = Base_data_copy.gender)
plt.title('Distribution of Gender')
```

```python
labels = [f'{i} -{i+9}' for i in range(0, 101, 10)]
print(labels)
```

```python
Base_data_copy['Math Group'] = pd.cut(Base_data_copy['math score'], range(0,111,10), labels = labels, 
                                      right = False)
```

```python
Base_data_copy.head()
```

```python
median_female_math_score = Base_data_copy[Base_data_copy['gender'] == 'female']['math score'].median()

print(median_female_math_score)
```

```python
Base_data_copy[Base_data_copy['gender'] == 'female']['math score'].describe()
```

```python
median_male_math_score = Base_data_copy[Base_data_copy['gender'] == 'male']['math score'].median()
print(median_male_math_score)
```

```python
Base_data_copy[Base_data_copy['gender'] == 'male']['math score'].describe()
```

```python
count_above_median = (Base_data_copy['math score'] >= median_male_math_score).sum()

print(f"Number of people who scored above the median: {count_above_median}")
```

```python
pd.crosstab(Base_data_copy.gender, Base_data_copy['Math Group'])
```

```python
sns.countplot(data = Base_data_copy, x = 'Math Group', hue = 'gender')
plt.xticks( fontsize = 7)
plt.title('Math Group Distribution by Gender')
```

```python
sns.kdeplot(data = Base_data_copy, x = 'math score', hue = 'gender', fill =True)
plt.axvline(median_male_math_score, color='orange', linestyle='--', label=f'Median: {median_male_math_score:.2f}')
plt.axvline(median_female_math_score, color='cyan', linestyle='--', label=f'Median: {median_female_math_score:.2f}')

from matplotlib.lines import Line2D
custom_lines = [
    Line2D([0], [0], color='orange', lw=2, linestyle='--', label=f'Male Median: {median_male_math_score:.2f}'),
    Line2D([0], [0], color='blue', lw=2, linestyle='--', label=f'Female Median: {median_female_math_score:.2f}')
]

# Show combined legend
plt.legend(handles=custom_lines, title='Medians')
plt.title('Math Score Distribution by Gender')
plt.xlabel('Math Score')
plt.ylabel('Density')
plt.show()
```

```python

# Using seaborn to create the grouped bar chart
sns.barplot(data=Base_data_copy, 
            x='race/ethnicity',  # X-axis: Race/Ethnicity groups
            y='math score',      # Y-axis: Math Scores
            hue='gender',        # Grouped by Gender (e.g., Male, Female)
            ci=None,             # No confidence interval bars
            palette='Set1')      # Color palette for distinction

# Customizing the chart
plt.title('Math Scores by Race/Ethnicity and Gender')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Mean Math Score')
plt.xticks(rotation=45, ha='right')  # Rotates X labels for better readability
plt.legend(title='Gender', loc='upper left')

# Show the plot
plt.tight_layout()  # Adjusts the plot for a better layout
plt.show()
```

```python
# Pivoting the data to create a matrix for heatmap
pivot_data = Base_data_copy.pivot_table(columns='parental level of education', 
                            index='race/ethnicity', 
                            values='math score', 
                            aggfunc='median')

# Creating a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, 
            annot=True,  # To display the values in each cell
            cmap='coolwarm',  # Color map
            fmt='.1f',  # Format of numbers in the cells
            linewidths=0.5,  # Border thickness between cells
            cbar_kws={'label': 'Mean Math Score'})  # Colorbar label

# Customizing the plot
plt.title('Median Math Scores by parental level of education and race/ethnicity')
plt.ylabel('race/ethnicity')
plt.xlabel('parental level of education')
plt.xticks(rotation = 360, ha = 'center', fontsize = 9)

plt.savefig('heatmap_math_scores.png', dpi=400) 
# Show the plot
plt.tight_layout()
plt.show()

```

```python
dummy_df = Base_data_copy.copy()
```

```python
dummy_df = dummy_df.drop(['reading score','writing score','Average'], axis = 1)
```

```python
dummy_df['Math pass'] = (dummy_df['math score'] > 66).astype(int)
```

```python
dummy_df = dummy_df.drop(['math score'], axis = 1)
```

```python
dummy_data = pd.get_dummies(dummy_df).astype(int)
```

```python
dummy_data.head()
```

```python
plt.figure(figsize=(20,10))
plt.title('Correlation between individual variable and Math pass variable')

dummy_data.corr(numeric_only =True)['Math pass'].sort_values(ascending=False).plot(kind='bar')

plt.tight_layout()
plt.savefig('Correlation of math scores and individual variables',dpi=400)
plt.show()
```
